```python
# sp500 stock data last 5-years
sp500 = yf.Ticker('^GSPC')
sp500_hist = sp500.history(period='5y')
sp500_hist.to_csv("sp500_hist.csv")
sp500_hist
```

```python
#crypto URL's for real-time data, if needed
#I can pull out price, supply, marketcap, etc. if we want to use this data
btc_url = "https://api.alternative.me/v2/ticker/Bitcoin/?convert=CAD"
eth_url = "https://api.alternative.me/v2/ticker/Ethereum/?convert=CAD"
dash_url = "https://api.alternative.me/v2/ticker/dash/?convert=CAD"
litecoin_url = "https://api.alternative.me/v2/ticker/litecoin/?convert=CAD"
bitcoin_cash_url = "https://api.alternative.me/v2/ticker/bitcoin-cash/?convert=CAD"
```

```python
#bitcoin real-time data
btc_response = requests.get(btc_url)
btc_json=btc_response.json()
btc_data = pd.json_normalize(btc_json['data'])
btc_data.to_csv("btc_data.csv")
btc_data
```

```python
#ethereum real-time data
eth_response = requests.get(eth_url)
eth_json=eth_response.json()
eth_data = pd.json_normalize(eth_json['data'])
eth_data.to_csv("eth_data.csv")
eth_data
```

```python
#dash real-time data
dash_response = requests.get(dash_url)
dash_json=dash_response.json()
dash_data = pd.json_normalize(dash_json['data'])
dash_data.to_csv("dash_data.csv")
dash_data
```

```python
#litcoin real-time data
litcoin_response = requests.get(litecoin_url)
litcoin_json=litcoin_response.json()
litcoin_data = pd.json_normalize(litcoin_json['data'])
litcoin_data.to_csv("litcoin_data.csv")
litcoin_data
```

```python
#bitcoin cash real-time data
bitcoin_cash_response = requests.get(bitcoin_cash_url)
bitcoin_cash_json=bitcoin_cash_response.json()
bitcoin_cash_data = pd.json_normalize(bitcoin_cash_json['data'])
bitcoin_cash_data.to_csv("bitcoin_cash_data.csv")
bitcoin_cash_data
```
